```python
import numpy as np
```

```python
original = []
transliterated = []

with open("125-150_test.md", "r") as fp:
    for i,line in enumerate(fp):
        if i%2 == 0:
            original.append(line)
        else:
            transliterated.append(line)
```

```python
parallel = []
for i,j in zip(original,transliterated):
    words_i = i.split()
    words_j = j.split()
    
    for wordi, wordj in zip(words_i,words_j):
        parallel.append((wordi, wordj))
```

```python
for item in set(parallel):
    print(item[0],item[1])
```
