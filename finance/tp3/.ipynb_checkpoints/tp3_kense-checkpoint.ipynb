```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import time
%matplotlib inline
```

```python
#read data from file
data = pd.read_table("./EUR_USD-20120101-20120301.dat", header=None, names=["timestamp", "bid", "ask"])
data["timestamp"] = data["timestamp"].apply(lambda x : time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime(x)))
print(data.head())
print(len(data))
```

```python
#plot test
plt.plot("timestamp","bid",data=data.head(1000))
plt.xticks([])
plt.title("EUR/USD Exchange Rate")
plt.xlabel("Timestamp")
plt.ylabel("Bid")
plt.show()
```
