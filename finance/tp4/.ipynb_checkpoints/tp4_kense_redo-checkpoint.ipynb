### Time Average

```python
#artificial time series
np.random.seed(1)
x = 100 + np.cumsum(0.5-np.random.random(1000))
```

```python
#visualize the time series
plt.plot(x)
plt.xlabel("time")
plt.ylabel("value")
plt.title("Artificial Times Series")
```

```python
#Moving Average for N = 100
N = 100

def MA(ts, N):
    L = len(ts)
    assert L%N == 0 #we want to make sure even blocks
    ma = []
    
    for i in range(0,N): #account for necessary data
        ma.append(np.nan)
        
    for i in range(N,L): #for each data point starting from N
        ma.append(np.ma.average(ts[i:i+N]))
    
    return ma
```

```python
plt.plot(x, label = "Series")
plt.xlabel("time")
plt.ylabel("value")
plt.title("Artificial Times Series with MA")
plt.plot(MA(x,N), label="MA")
plt.legend()
plt.show()
```

```python
def EMA(ts, N, smoothing):
    L = len(ts)
    assert L%N == 0 #we want to make sure even blocks
    ema_init = np.ma.average(ts[N:N+N]) #init with SMA
    
    ema = []
    for i in range(0,N): #lag
        ema.append(np.nan)
        
    ema_pre = ema_init #first run
    for i in range(N, L):
        ema_curr = ts[i]*(smoothing) + ema_pre*(1-smoothing)
        ema.append(ema_curr)
        ema_pre = ema_curr
    
    return ema
```

```python
plt.plot(x, label = "Series")
plt.xlabel("time")
plt.ylabel("value")
plt.title("Artificial Times Series with EMA, smoothing = 2/(100+1)")
plt.plot(EMA(x,N,2/(100+1)), label="EMA")
plt.legend()
plt.show()
```

```python
plt.plot(x, label = "Series")
plt.xlabel("time")
plt.ylabel("value")
plt.title("Artificial Times Series with EMA, smoothing = 0.1")
plt.plot(EMA(x,N,0.1), label="EMA")
plt.legend()
plt.show()
```

```python
plt.plot(x, label = "Series")
plt.xlabel("time")
plt.ylabel("value")
plt.title("Artificial Times Series with EMA, smoothing = 0.01")
plt.plot(EMA(x,N,0.01), label="EMA")
plt.legend()
plt.show()
```

In the case of smoothing `smoothing = 0.1` and `smoothing = 0.01`, we can see that a smaller value (meaning more smoothing) means that prices are less affected by sudden shifts compared to the higher value.0.019801980198019802
For the case of 2/(N+1), we can see that `2/(N+1) = 0.019801980198019802` and so we should get a smoothing somewhere between 0.1 and 0.01, and that reflects the shape of the data somewhat.
