```python
P1 = data[data.participant == 'D02P01']
P2 = data[data.participant == 'D02P02']
```

```python
#event log tracing P1
trace_P1 = []
for i in range(len(P1)): #trace over all events of P1
    
    #if first run
    if i == 0:
        trace_P1.append(P1.iloc[i])

    else:
        #check previous before adding
        if P1.iloc[i-1]['collaborative.processes'] == P1.iloc[i]['collaborative.processes']:
            continue
        
        #otherwise add
        else:
            trace_P1.append(P1.iloc[i])

#event log tracing function
def trace(eventlog):
    trace_ = []
    for i in range(len(eventlog)):
        
        #if fun run
        if i == 0:
            trace_.append(eventlog.iloc[i])
        else:
            #check previous
            if eventlog.iloc[i-1]['collaborative.processes'] == eventlog.iloc[i]['collaborative.processes']:
                continue
            
            #otherwise add
            else:
                trace_.append(eventlog.iloc[i])
    return trace_
```

### Explanation
The reasoning here for going through the collaborative process is that the first primary assumption that we're making is that since it is a collaborative exchange, it is going to be natural that the two persons will take turns interacting. Therefore, we first go beyond this layer of abstraction to see the individual actions (which we know to be influenced by the other) as if they are intrinsic to the singular actors (independence assumption to some extent).
Then, we look at each collaborative process, and add ones that are unique to at least the next immediate action, which is an assumption that removes some duplicity.
Thus, we can get a "log" replay or trace of some of the actions the individuals took.
We can then explore that.

```python
trace_P1 = pd.DataFrame(data=trace_P1)
trace_P2 = pd.DataFrame(data=trace(P2))
```

```python
#lets see how many times each process is followed by "relax atmosphere"
relax_followup = {}
for i in range(len(trace_P1)-1):
    
    #if first encounter
    if trace_P1.iloc[i]['collaborative.processes'] not in relax_followup:
        relax_followup[trace_P1.iloc[i]['collaborative.processes']] = 0
        
    #if next is relax atmosphere
    if trace_P1.iloc[i+1]['collaborative.processes'] == 'relax.atmosphere':
        relax_followup[trace_P1.iloc[i]['collaborative.processes']] += 1
    else:
        continue
```

```python
relax_followup
```

```python
#now we can do something like, find the reverse
relax_count = {}
for i in range(len(trace_P1)-1):
        
    #first encounter
    if trace_P1.iloc[i]['collaborative.processes'] not in relax_count:
        relax_count[trace_P1.iloc[i]['collaborative.processes']] = 0
    
    #if current is relax atmosphere    
    if trace_P1.iloc[i]['collaborative.processes'] == 'relax.atmosphere':
        
        #if first encounter
        if trace_P1.iloc[i+1]['collaborative.processes'] not in relax_count:
            relax_count[trace_P1.iloc[i+1]['collaborative.processes']] = 0
        
        #then check the followup
        relax_count[trace_P1.iloc[i+1]['collaborative.processes']] += 1
```

```python
#now we can evaluate if activity is followed by relax atmosphere, but never following relax atmosphere
causal = []
for activity in relax_followup:
    if relax_followup[activity] > 0 and relax_count[activity] == 0:
        causal.append(activity)
```

### Explanation
Here, we can see that there are some dependencies in the ways of the a-algorithm. We can now apply this kind of process to the entire trace of the individual P1 and P2 logs.

```python
def dependency_causality (eventlog):
    '''
    find the dependencies in eventlog
    '''
    
    #create each count
    dependencies = {}
    '''
    dependencies is a dictionary of dictionaries {a:{b}}, where each entry has their own counts,
    whereby the counts are the number of times b follows a (a->b)
    '''
    
    causalities = {}
    '''
    causalities is a dictionary showing the frequencies
    '''
    
    
    #cycle through the eventlog
    for i in range(len(eventlog)):
        
        #if first run create the process dependency
        if eventlog.iloc[i]['collaborative.processes'] not in dependencies:
            dependencies[eventlog.iloc[i]['collaborative.processes']] = {}
            
            for j in range(len(eventlog)):
                #not dupe
                if eventlog.iloc[j]['collaborative.processes'] != eventlog.iloc[i]['collaborative.processes']:
                    dependencies[eventlog.iloc[i]['collaborative.processes']][eventlog.iloc[j]['collaborative.processes']] = 0
            
        #create dependencies
        if i < len(eventlog)-1:
            dependencies[eventlog.iloc[i]['collaborative.processes']][eventlog.iloc[i+1]['collaborative.processes']] += 1

    #determine all causalities
    for entry in dependencies:
        for activity in dependencies[entry]:
            if dependencies[entry][activity] > 0 and dependencies[activity][entry] == 0:
                
                #add causality or increase count
                if "{0}->{1}".format(entry,activity) not in causalities:
                    causalities["{0}->{1}".format(entry,activity)] = 1
                else:
                    causalities["{0}->{1}".format(entry,activity)] += 1
    
    return dependencies, causalities
```

```python
dependencies_P1, causalities_P1 = dependency_causality(trace_P1)
dependencies_P2, causalities_P2 = dependency_causality(trace_P2)
```
